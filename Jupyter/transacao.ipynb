```python
def get(self, transacao_id):
        for transacao in transacoes:
            if transacao['id'] == transacao_id:
                return transacao, 200
        return "Transação não encontrada", 404

    def post(self):
        global id_transacao
        parser = reqparse.RequestParser()
        parser.add_argument("usuario_id", type=int, required=True, help="ID do usuário para qual a transação será registrada")
        parser.add_argument("tipo", type=str, required=True, help="Tipo da transação (depósito ou retirada)")
        parser.add_argument("valor", type=float, required=True, help="Valor da transação")
        args = parser.parse_args()

        transacao = {
            "id": id_transacao,
            "usuario_id": args["usuario_id"],
            "tipo": args["tipo"],
            "valor": args["valor"],
            "data": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        id_transacao += 1
        transacoes.append(transacao)
        return transacao, 201

    def put(self, transacao_id):
        parser = reqparse.RequestParser()
        parser.add_argument("usuario_id", type=int, required=True, help="ID do usuário para qual a transação será registrada")
        parser.add_argument("tipo", type=str, required=True, help="Tipo da transação (depósito ou retirada)")
        parser.add_argument("valor", type=float, required=True, help="Valor da transação")
        args = parser.parse_args()

        for transacao in transacoes:
            if transacao['id'] == transacao_id:
                transacao['usuario_id'] = args["usuario_id"]
                transacao['tipo'] = args["tipo"]
                transacao['valor'] = args["valor"]
                transacao['data'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                return transacao, 200

        transacao = {
            "id": transacao_id,
            "usuario_id": args["usuario_id"],
            "tipo": args["tipo"],
            "valor": args["valor"],
            "data": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        transacoes.append(transacao)
        return transacao, 201

    def delete(self, transacao_id):
        global transacoes
        transacoes = [transacao for transacao in transacoes if transacao['id'] != transacao_id]
        return "{} foi deletada.".format(transacao_id), 200
```
